```python
import requests
movie_name = input('Enter Movie Name :')
movielist = list()

api_key = "72a0b3621616ed0aecda44398d8371f9"
api_link = "https://api.themoviedb.org/3"
params = {'query':movie_name, 'api_key':api_key}
header = {'Accept': 'application/json'}


response = requests.get(api_link + "/search/movie", headers = header, params=params)
data = response.json()
results = data.get('results')
for result in results:
    if result.get('title') == movie_name:
        id = result.get('id')
        
params2 = {'api_key':api_key}
response2 = requests.get(api_link + "/movie/" + str(id) + "/similar", headers = header, params=params2)
data2 = response2.json()
results2 = data2.get('results')
for result in results2[:6]:
    movie = result.get("title")
    movielist.append(movie)

if len(movielist) >= 1:
    for movie in movielist:
        print(movie)
else:
    print('Sorry!! Movie name is incorrect')
```

```python
import requests
regionname = input('Enter Region Code :')

link = 'https://api.themoviedb.org/3'
key = '72a0b3621616ed0aecda44398d8371f9'
param = {'api_key':key , 'region':regionname}
header = {'Accept': 'application/json'} 
response = requests.get(link+'/movie/top_rated', headers = header , params = param)
response.status_code
data = response.json()
results = data['results']

title_array = []
genre_id_array = []

for result in results:
    if result.get('original_language') == 'en':
        title_array.append(result.get('title'))
        genre_id_array.append(result.get('genre_ids'))

response2 = requests.get(link + "/genre/movie/list", headers = header, params = param)
data2 = response2.json()
genres = data2.get('genres')
mapping = {}

for genre in genres:
    mapping[genre.get('id')] = genre.get('name')

for i in range(10):
    print(title_array[i], "-", end=" ")
    for id in genre_id_array[i]:
        print(mapping.get(id), end = ", ")
    print()
```
